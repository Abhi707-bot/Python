```python
#Import pandas and csv

import pandas as pd

df = pd.read_csv("CricketMatchData.csv")
```

```python
df
```

```python
#Rename columns

# Replace column names with the first row
df.columns = df.iloc[0]


# Drop the first row (as it's now the header)
df = df[1:].reset_index(drop=True)
```

```python
df
```

```python
#Rename columns in a list

df = df.rename(columns = {'Mat':'Matches','NO':'Not_Outs', 'HS':'Highest_Innings_Score', 'BF':'Balls_Faced', 'SR':'Batting_Strike_Rate'})
```

```python
df.head()
```

```python
#Check the NUlls 

df.isnull().any()
```

```python
df[df['Balls_Faced'].isna()==1]
```

```python
df['Balls_Faced'] = df['Balls_Faced'].fillna(0)
```

```python
df['Batting_Strike_Rate'] = df['Batting_Strike_Rate'].fillna(0)
```

```python
#Drop Duplicates

df.duplicated()
```

```python
df = df.drop_duplicates()
```

```python
#Split up Span into start and end date

df['Rookie_Year'] = df['Span'].astype(str).str[4:]

df['Final_Year'] = df['Span'].astype(str).str[:4]

df.head()
```

```python
df = df.drop(columns=['Span'])
```

```python
df.head()
```

```python
#Split up Contry code from players

df['Player'].str.split(pat = '(')
```

```python
df['Country'] = df['Player'].str.split(pat = '(').str[1]
```

```python
df['Country'] = df['Country'].str.split(pat = ')').str[0]
```

```python
df['Country']
```

```python
df['Player'] = df['Player'].str.split(pat = '(').str[0]
```

```python
df.head()
```

```python
df['Highest_Innings_Score'] = df['Highest_Innings_Score'].str.split(pat = '*').str[0]
```

```python
df['Highest_Innings_Score']
```

```python
df.head()
```

```python
#Change DataTypes

df['Highest_Innings_Score'] = df['Highest_Innings_Score'].astype('int')
```

```python
df.dtypes
```

```python
df = df.astype({'Rookie_Year':'int', 'Final_Year':'int'})
```

```python
df.dtypes
```

```python
df['Matches'] = df['Matches'].str.split(pat = '*').str[0]
```

```python
df['Matches'] = df['Matches'].astype('int')
```

```python
df.dtypes
```

```python
df['Balls_Faced'] = df['Balls_Faced'].str.split(pat = '+').str[0]
```

```python
df['Balls_Faced'] = df['Balls_Faced'].astype('int')
```

```python
df['Batting_Strike_Rate'] = df['Batting_Strike_Rate'].astype('float')
```

```python
#Build out career length (hint: subract rookie and final year)

df['Career_Length'] = df['Rookie_Year'] - df['Final_Year']
```

```python
df.head()
```

```python
# Calculatins with the dataframe
```

```python
# Ques 1) --> in this df whats avg career length?

df['Career_Length'].mean()
```

```python
# Ques 2) --> Avg batting strike rate for cricketers who played over 10 yrs

df[df['Career_Length']>10]['Batting_Strike_Rate'].mean()
```

```python
# Ques 3) --> find cric who played before 1960s

df[df['Rookie_Year']<1960]['Player'].count()
```

```python
# Ques 4) --> Max highest inn score by country

df.groupby('Country')['Highest_Innings_Score'].max().to_frame('Inn_By_Country').reset_index().sort_values('Inn_By_Country', ascending = False)
```

```python
# Ques 5) --> Hundreds, Fifties, ducks avg by country

df.groupby('Country')[[100, 50, 0.0]].mean()
```
